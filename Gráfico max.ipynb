```python
import pandas as pd
from matplotlib import pyplot as plt    
import numpy as np  
```

```python
germ = pd.read_csv("germination.csv")
germ
```

```python
germ= germ.groupby(["species","temp"], sort=False) ["germinated"].max()
germ   
```

```python
ax= germ.plot(kind= "bar", subplots= True)
plt.show()   
```

```python
ax= germ.plot(kind= "bar", subplots= False)
ax.set_title("Maximum germination per temperature")
ax.set_xlabel('Temperatura')
ax.set_ylabel('GerminaciÃ³n')
plt.show()   
```

```python
# ComparaciÃ³n mayor germinaciÃ³n entre las especies en funciÃ³n de los diferentes temperatura 
germ.unstack().T.plot(kind="bar", subplots=False, layout=(2,2),figsize=(20,10), width=1) 
plt.show()  
# Necesito poner titulo al eje y como :GerminaciÃ³n
# Y titulo de la grÃ¡fica : CamparaciÃ³n de mayor germinaciÃ³n
```

```python
germ.unstack().T.plot.bar(subplots=False, layout=(2,2),figsize=(20,10), width=1) 
plt.show()  
```
