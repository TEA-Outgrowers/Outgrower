```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("germination.csv")
data
```

```python
data = data.groupby(["species","temp","end"], sort=False)["germinated"].max()
```

```python
data.unstack().T.plot(kind="bar", subplots=False, layout=(2,2),figsize=(20,10), width=1)
```
