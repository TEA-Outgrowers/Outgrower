```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("germination.csv")
data
```

```python
# rice data 
rice = data[data.species == "rice"]
rice
```

```python
# wheat data 
wheat = data[data.species == "wheat"]
wheat
```

```python
# wheat temp 10
wheat.iloc[0:17]    
```

```python
# total wheat germinated at temp 10
w = wheat.germinated.iloc[0:17] 
w 
w.sum()   
```

```python
# combination at temp 10
start = wheat.temp.iloc[0:17]
end = wheat.end.iloc[0:17]
germ = wheat.germinated.iloc[0:17]
x = germ.combine(start) 
```

```python
# mungbean data 
mungbean = data[data.species == "mungbean"]
mungbean   
```

```python
# mungbean at temp 10
mungbean.loc[17:24]      
```

```python
# total mungbean germinated at temp 10
m = mungbean.germinated.loc[17:24] 
m    
m.sum()   
```

```python
# comparison temp and germinated for wheat and mungbean
plt.plot(wheat.temp, wheat.germinated)
plt.plot(mungbean.temp, mungbean.germinated)
plt.legend(["wheat", "mungbean"])
plt.xlabel("Temperatura")
plt.ylabel("Germinacion")
plt.show()    
```
